# Supervised learning on cleaned data

- Random Forest
- Decision Trees
- Gradient Boosting
- SVM
- Naive Bayes
- AdaBoost

First test the models on the 2003 data for different (balanced) train sizes. Next, use the optimal models (with optimal trian size) to test their generalizability on the years 2007, 2010, 2012, 2014, 2018.

```python
# Load libraries
import numpy as np
import pandas as pd
from plotnine import *
from sklearn.svm import SVC
from collections import Counter
from imblearn.pipeline import Pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
```

```python
# Load data
df = pd.read_csv('data/2003/clean_labeled_zero_2003.csv')
df_sub = df[['BC', 'ID', 'OD', 'CC', 'DC', 'Type']]
X = df_sub.drop(columns=['Type'])
y = df_sub['Type']
```

```python
%%capture
# Create model pipelines
pip_dt = Pipeline([('under', RandomUnderSampler(random_state=42)), ('model', DecisionTreeClassifier(random_state=42))])
pip_rf = Pipeline([('under', RandomUnderSampler(random_state=42)), ('model', RandomForestClassifier(n_estimators=100, random_state=42))])
pip_gb = Pipeline([('under', RandomUnderSampler(random_state=42)), ('model', GradientBoostingClassifier(n_estimators=100, random_state=42))])
pip_svm = Pipeline([('under', RandomUnderSampler(random_state=42)), ('model', SVC(kernel='poly', degree=4, random_state=42))])
pip_nb = Pipeline([('under', RandomUnderSampler(random_state=42)), ('model', GaussianNB())])
pip_ada = Pipeline([('under', RandomUnderSampler(random_state=42)), ('model', AdaBoostClassifier(n_estimators=100, random_state=42))])

models = [(pip_dt, "dt"), (pip_rf, "rf"), (pip_gb, "gb"), (pip_svm, "svm"), (pip_nb, "nb"), (pip_ada, "ada")]

label_0 = []
label_1 = []
label_2 = []
label_3 = []

x = sum(df['Type'].value_counts() // df['Type'].value_counts().min())
table_x = [x * i for i in range(1, 95, 4)]
np.insert(table_x, 0, 1)

for model in models:
    for i in table_x:
        X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=i, random_state=42, stratify=y)
        undersampler = RandomUnderSampler(random_state=42)
        X_under, y_under = undersampler.fit_resample(X_train, y_train)
        model[0].fit(X_under, y_under)
        y_pred = model[0].predict(X_test)
        report = classification_report(y_test, y_pred, output_dict=True)
        label_0.append((model[1], i, 0, report['0']['f1-score']))
        label_1.append((model[1], i, 1, report['1']['f1-score']))
        label_2.append((model[1], i, 2, report['2']['f1-score']))
        label_3.append((model[1], i, 3, report['3']['f1-score']))
```

```python
scores = [*label_0, *label_1, *label_2, *label_3]
model, n_rows, label, f1_score = zip(*scores)
```

```python
data = {'model': model, 
        'n_rows': n_rows,
        'label': label, 
        'f1-score': f1_score
       }

df1 = pd.DataFrame(data)
```

```python
df1.to_csv("data/train_results_2003.csv", index=False)
```

## Test the models on 2007, 2010, 2012, 2014, 2018

```python
# Load data
df_2003 = pd.read_csv('data/2003/clean_labeled_zero_2003.csv')
df_sub_2003 = df_2003[['BC', 'ID', 'OD', 'CC', 'DC', 'Type']]
X_2003 = df_sub_2003.drop(columns=['Type'])
y_2003 = df_sub_2003['Type']
```

```python
# Load data
df_2018 = pd.read_csv('data/2018/clean_labeled_zero_2018.csv') # Alter depending on the test year
df_sub_2018 = df_2018[['BC', 'ID', 'OD', 'CC', 'DC', 'Type']]
X_2018 = df_sub_2018.drop(columns=['Type'])
y_2018 = df_sub_2018['Type']
```

```python
# Create model pipelines
pip_dt = Pipeline([('under', RandomUnderSampler(random_state=42)), ('model', DecisionTreeClassifier(random_state=42))])
pip_rf = Pipeline([('under', RandomUnderSampler(random_state=42)), ('model', RandomForestClassifier(n_estimators=100, random_state=42))])
pip_gb = Pipeline([('under', RandomUnderSampler(random_state=42)), ('model', GradientBoostingClassifier(n_estimators=100, random_state=42))])
pip_svm = Pipeline([('under', RandomUnderSampler(random_state=42)), ('model', SVC(kernel='poly', degree=4, random_state=42))])
pip_nb = Pipeline([('under', RandomUnderSampler(random_state=42)), ('model', GaussianNB())])

models = [(pip_dt, "dt", 3675), (pip_rf, "rf", 3675), (pip_gb, "gb", 3225), (pip_svm, "svm", 4875), (pip_nb, "nb", 375)]

label_0 = []
label_1 = []
label_2 = []
label_3 = []

for model in models:
    X_train, X_test, y_train, y_test = train_test_split(X_2003, y_2003, train_size=model[2], random_state=42, stratify=y_2003)
    model[0].fit(X_train, y_train)
    y_pred = model[0].predict(X_2018)
    report = classification_report(y_2018, y_pred, output_dict=True)
    label_0.append((model[1], model[2], "0", report['0']['f1-score'], report['0']['precision'], report['0']['recall']))
    label_1.append((model[1], model[2], "1", report['1']['f1-score'], report['1']['precision'], report['1']['recall']))
    label_2.append((model[1], model[2], "2", report['2']['f1-score'], report['2']['precision'], report['2']['recall']))
    label_3.append((model[1], model[2], "3", report['3']['f1-score'], report['3']['precision'], report['3']['recall']))
```

```python
scores = [*label_0, *label_1, *label_2, *label_3]
model, n_rows, label, f1_score, precision, recall = zip(*scores)
```

```python
data = {'model': model, 
        'n_rows': n_rows,
        'label': label,
        'f1-score': f1_score,
        'precision': precision,
        'recall': recall
       }

df1 = pd.DataFrame(data)

df1.sort_values(by=['model', 'label'])
```

```python
df1.to_csv("gen_test_2007.csv", index=False) # Alter depending on the test year
```

```python
df1 = pd.read_csv("gen_test_2007.csv").assign(year=lambda x: 2007)
df2 = pd.read_csv("gen_test_2010.csv").assign(year=lambda x: 2010)
df3 = pd.read_csv("gen_test_2012.csv").assign(year=lambda x: 2012)
df4 = pd.read_csv("gen_test_2014.csv").assign(year=lambda x: 2014)
df5 = pd.read_csv("gen_test_2018.csv").assign(year=lambda x: 2018)
```
